# Shape

---
统计图片wh分布

```python
image_dir = './up-3d-box/'
im_names = os.listdir(image_dir)
heights = []
weights = []
dict_heights = {}
dict_weights = {}
for i,name in enumerate(im_names):
    print(i+1,'/',len(im_names),end='\r')
    im_path = os.path.join(image_dir,name)
    im = cv2.imread(im_path)
    if im is None:
        print(name)
        continue
    height,weight,_= im.shape
    heights.append(height)
    weights.append(weight)
    if height not in dict_heights.keys():
        dict_heights[height] = 0
    if weight not in dict_weights.keys():
        dict_weights[weight] = 0
    dict_heights[height] += 1
    dict_weights[weight] += 1
```

```python
heights_keys = list(dict_heights.keys())
heights_values = list(dict_heights.values())
weights_keys = list(dict_weights.keys())
weights_values = list(dict_weights.values())
plt.figure(1,figsize=(20,20))
plt.title('height')
plt.bar(left=heights_keys,height=heights_values,color="lightblue")
plt.figure(2,figsize=(20,20))
plt.title('weight')
plt.bar(left=weights_keys,height=weights_values,color="lightblue")
```

```python
plt.figure(1,figsize=(20,20))
plt.title('h-w')
plt.scatter(heights,weights,color="lightblue")
plt.xlabel('heights')
plt.ylabel('weights')
```

# Class number

---
统计按bmi的胖瘦分类分布  
0 偏瘦	<= 18.4  
1 正常	18.5 ~ 23.9   
2 过重	24.0 ~ 27.9  
3 肥胖	>= 28.0

```python
f = open('./women_v1_train.json','r')
info_1 = json.load(f)
f = open('./women_v1_val.json','r')
info_2 = json.load(f)
```

```python
class_ids_dict = {0:u"Thin",1:u"Nomal",2:u"Fat",3:u"Fater"}
```

```python
len(info_1)
```

```python
len(info_2)
```

```python
class_ids_1 = {}
for key in info_1.keys():
    class_id = info_1[key]['label']
    class_id = class_ids_dict[class_id]
    if class_id not in class_ids_1:
        class_ids_1[class_id] = 0
    class_ids_1[class_id] += 1

class_ids_2 = {}
for key in info_2.keys():
    class_id = info_2[key]['label']
    class_id = class_ids_dict[class_id]
    if class_id not in class_ids_2:
        class_ids_2[class_id] = 0
    class_ids_2[class_id] += 1
```

```python
keys_1 = list(class_ids_1.keys())
keys_2 = list(class_ids_2.keys())
values_1 = list(class_ids_1.values())
values_2 = list(class_ids_2.values())
```

```python
plt.figure(1)
plt.title('info_1 label')
a = plt.bar(left=keys_1,height=values_1,color="lightblue")
autolabel(a)
plt.figure(2)
plt.title('info_2 label')
a = plt.bar(left=keys_2,height=values_2,color="lightblue")
autolabel(a)
```

# Image Mean

---
计算图片均值

```python
im_dir = '../dataset/bodytype/sample/'
names = os.listdir(im_dir)
```

```python
mean_1 = []
mean_2 = []
mean_3 = []
for name in names:
    path = os.path.join(im_dir,name)
    im = cv2.imread(path)
    mean_1.append(np.mean(im[:,:,0]))
    mean_2.append(np.mean(im[:,:,1]))
    mean_3.append(np.mean(im[:,:,2]))
```

```python
f = open('text.txt','w+')
```

```python
shape = np.shape([mean_1,mean_2,mean_3])
```

```python
f.write(str(shape)+'\n')
```

```python
mean = np.mean([mean_1,mean_2,mean_3],axis=1)
```

```python
print(mean)
```

```python
f.write(str(mean.tolist())+'\n')
```

# The number of Class  by shape

---
按照体型分类的分布统计  
'yHourglass':0  
'yPear':1  
'yApple':2  
'yBanana':3  

```python
# shape_dict = {'yHourglass':0,'yPear':1,'yApple':2,'yBanana':3,'yEmpty':4}
# class_ids_dict = {0:u"yHourglass",1:u"yPear",2:u"yApple",3:u"yBanana",4:u"yEmpty"}
shape_dict = {'yHourglass':0,'yPear':1,'yApple':2,'yBanana':3}
class_ids_dict = {0:u"yHourglass",1:u"yPear",2:u"yApple",3:u"yBanana"}
```

```python
f = open('./women_v4_train.json','r')
train_info = json.load(f)
f = open('./women_v4_val.json','r')
val_info = json.load(f)
```

```python
train_class_ids = {}
for key in train_info.keys():
    class_id = train_info[key]['label']
    class_id = class_ids_dict[class_id]
    if class_id not in train_class_ids:
        train_class_ids[class_id] = 0
    train_class_ids[class_id] += 1

val_class_ids = {}
for key in val_info.keys():
    class_id = val_info[key]['label']
    class_id = class_ids_dict[class_id]
    if class_id not in val_class_ids:
        val_class_ids[class_id] = 0
    val_class_ids[class_id] += 1
```

```python
train_label_keys = list(train_class_ids.keys())
val_label_keys = list(val_class_ids.keys())
train_label_values = list(train_class_ids.values())
val_label_values = list(val_class_ids.values())
```

```python
plt.figure(1)
plt.title('train label')
a = plt.bar(left=train_label_keys,height=train_label_values,color="lightblue")
autolabel(a)
plt.figure(2)
plt.title('val label')
a = plt.bar(left=val_label_keys,height=val_label_values,color="lightblue")
autolabel(a)
```
