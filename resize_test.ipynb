```python
import cv2
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
im_dir = './dataset/image_women/'
name_list = os.listdir(im_dir)
INPUT_SHAPE = (1024,512,3)
MEAN_PIXEL = np.array([93.2,104.6,116.6])
```

```python
path = os.path.join(im_dir,name_list[15])
im_pad = np.zeros(INPUT_SHAPE,dtype=np.float64)
im = cv2.imread(path)
#im = im/255
im = im.astype(np.float64) - MEAN_PIXEL
```

```python
h,w = im.shape[:2]
print(h,w)
```

```python
if h/w > 2:
    re_h = INPUT_SHAPE[0]
    re_w = int(w * (re_h / h))
else:
    re_w = INPUT_SHAPE[1]
    re_h = int(h * (re_w / w))
print(re_h,re_w)
re_im = cv2.resize(im,(re_w,re_h))
im_pad[:re_h,:re_w,:] = re_im.copy()
print(re_im.shape,im_pad.shape)
```

```python
plt.figure(1)
plt.imshow(im)
plt.show()
plt.figure(2)
plt.imshow(re_im)
plt.show()
plt.figure(3)
plt.imshow(im_pad)
plt.show()
```
